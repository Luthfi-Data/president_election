### tweet preprocessing

```python
# remove URL
df_amin_tweet['cleaned_text'] = df_amin_tweet['full_text'].str.replace('https\S+', '', case=False)

# coverting to lowercase
df_amin_tweet['cleaned_text'] = df_amin_tweet['cleaned_text'].str.lower()

# replace all non-alphabetic characters with a space
df_amin_tweet['cleaned_text'] = df_amin_tweet['cleaned_text'].str.replace('[^a-zA-Z]', ' ', regex=True)

#remove words with len < 2
df_amin_tweet['cleaned_text'] = df_amin_tweet['cleaned_text'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
```

```python
df_amin_tweet.head(5)
```

### remove stop words

```python
tambahan = ['anies', 'prabowo', 'ganjar', 'imin', 'muhaimin', 'gibran', 'mahfud', 'md',
            'aniesmuhaimin', 'messi', 'irak', 'nadeo', 'hindia', 'rcti', 'you', 'jumatberkah',
            'adam alis', 'klok', 'arhan', 'asnawi', 'qory', 'tni', 'coldplay', 'amp', 'soobin',
            'sehun', 'shayne', 'jeketiselcaday', 'zhaolusi', 'black friday', 'coty', 'iraq',
            'dendy', 'black', 'alis', 'adam', 'pendek', 'friday', 'cak', 'subianto', 'prabowogemoy', 'prabowogibranistimewa',
            'gus', 'nomor', 'gak', 'aniesbaswedan', 'baswedan', 'cakiminow', 'persen', 'aminno']

stop_words = set().union(stopwords.words('indonesian'), tambahan)

df_amin_tweet['cleaned_text'] = [' '.join([word for word in x.lower().split() if word
                                                not in stop_words])
                                                for x in df_amin_tweet['cleaned_text'].tolist()]
```

### tokenize tweet

```python
# import word tokenize from nltk
tokenized_tweet = df_amin_tweet['cleaned_text'].apply(lambda x:
                  list(ngrams(x.split(), 1)))
```

```python
tokenized_tweet
```

### count most mentioned words

```python
# Pada baris ini, zip(tokenized_tweet) mengelompokkan elemen-elemen yang setara dari tokenized_tweet ke dalam tupel.
# Kemudian, reduce digunakan untuk menggabungkan semua tupel tersebut menjadi satu daftar (list(x) + list(y) berarti menggabungkan dua list).
# Hasilnya adalah satu daftar yang berisi semua kata dari tokenized_tweet
l = reduce(lambda x, y: list(x)+list(y), zip(tokenized_tweet))

# Baris ini menggunakan list comprehension untuk 'membuka' atau 'mendatar' struktur yang bersarang (list(x)+list(y) dari sebelumnya) menjadi satu dimensi.
# Hasilnya adalah daftar yang berisi semua kata dari tokenized_tweet.
flatten = [item for sublist in l for item in sublist]

# Pada baris ini, Counter digunakan untuk menghitung frekuensi setiap kata dalam daftar flatten.
# Kemudian, most_common() digunakan untuk mengurutkan kata-kata berdasarkan frekuensinya, mengembalikan list tuple berisi kata dan frekuensinya.
counts = Counter(flatten).most_common()

# create data frame
df_commonphrase_1 = pd.DataFrame.from_records(counts, columns=['Words', 'Count'])

# join phrase in one strin
df_commonphrase_1['Words']= df_commonphrase_1['Words'].apply(lambda x: ' '.join([w for w in x]))
```

### count most mentioned words

```python
l = reduce(lambda x, y: list(x)+list(y), zip(tokenized_tweet_3))

# Baris ini menggunakan list comprehension untuk 'membuka' atau 'mendatar' struktur yang bersarang (list(x)+list(y) dari sebelumnya) menjadi satu dimensi.
# Hasilnya adalah daftar yang berisi semua kata dari tokenized_tweet.
flatten = [item for sublist in l for item in sublist]

# Pada baris ini, Counter digunakan untuk menghitung frekuensi setiap kata dalam daftar flatten.
# Kemudian, most_common() digunakan untuk mengurutkan kata-kata berdasarkan frekuensinya, mengembalikan list tuple berisi kata dan frekuensinya.
counts = Counter(flatten).most_common()

# create data frame
df_commonwords_3 = pd.DataFrame.from_records(counts, columns=['Words', 'Count'])

# join phrase in one strin
df_commonwords_3['Words']= df_commonwords_3['Words'].apply(lambda x: ' '.join([w for w in x]))
```
